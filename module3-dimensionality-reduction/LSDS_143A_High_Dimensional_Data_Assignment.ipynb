#### $p()$

First we write out the input vector as scalars and basis vectors:

\begin{align}
\\
\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} = x_1\begin{bmatrix} 1 \\ 0 \end{bmatrix} + x_2\begin{bmatrix} 0 \\ 1 \end{bmatrix}

Then we pass the unit vectors $\hat{i}$ and $\hat{j}$ into the function individually to separate out the transformations of each $x$ value:

\begin{align}
p(\begin{bmatrix} 1 \\ 0 \end{bmatrix}) = \begin{bmatrix} 1 \\ -1 \end{bmatrix}
\\
p(\begin{bmatrix} 0 \\ 1 \end{bmatrix}) = \begin{bmatrix} 3 \\ 2 \end{bmatrix}
\end{align}

Now that we can see the effect of the transformation on the unit vectors, we can use those vectors to create our transformation matrix:

\begin{align}
T_p = \begin{bmatrix} 1 & 3 \\ -1 & 2 \end{bmatrix}
\end{align}

#### $q()$

\begin{align}
q(\begin{bmatrix}x_1 \\ x_2 \\ x_3\end{bmatrix}) = \begin{bmatrix} 4x_1 + x_2 + 2x_3 \\2 x_2 - x_1 + 3x_3 \\ 5x_1 - 2x_3 + x_2  \end{bmatrix}

\end{align}

Then we pass the unit vectors $\hat{i}$ and $\hat{j}$ into the function individually to separate out the transformations of each $x$ value:

\begin{align}
p(\begin{bmatrix} 1 \\ 0 \\ 0 \end{bmatrix}) = \begin{bmatrix} 4 \\ -1 \\ 5 \end{bmatrix}

\begin{align}
T_q = \begin{bmatrix} 4 & 1 & 2 \\ -1 & 2 & 3 \\ 5 & 1 & -2 \end{bmatrix}

# Eigenvalues and Eigenvectors

## 5.1

In your own words, give an explanation for the intuition behind eigenvalues and eigenvectors.

> An eigenvector is a vector that stays on its original span through a transformation.

In the case of a 2d vector, that means the transformed vector stayed on the line or axes of the original vector. In the case of a 3d vector, the eigenvector can be thought of as the axis of rotation. That term makes more visual sense in three dimensions than two.

> The eigenvalue is a scalar indicating by how much the eigenvector is transformed along its span.

# The Curse of Dimensionality

## 6.1

What are some of the challenges of working with high dimensional spaces?

> More dimensions means more difficult to explore and visualize

Being able to boil down data into fewer dimensions means simple visualizations that will likely be easier for the lay-person to understand. Although it might be clever to use color and shapes and other variables to show higher dimensions, it is much more difficult to draw firm conclusions and use the visualization to communicate those conclusions.

> The benefits gained from adding additional dimensions are usually smaller every time

In the case of a very large dataset, including extra features in the analysis could cost significant extra processing power and/or time. That is without the guarantee that those extra features are beneficial enough to the analysis to warrant that extra cost.

> Additional dimensions cause data points to spread out, eventually to the point of losing their meaningful relationships

Spreading the data out gives particularly bad results in the case of algorithms / models that rely on the distances between datapoints.

> Risk of overfitting models

A low ratio of observations to features increases the likelihood of models overfitting the data. For example, if the ratio is 1:1, some models will simply be trained to each specific data point.
