# Vertical Line Test

## 1.1 Create two graphs, one that passes the vertical line test and one that does not.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,2*(np.pi), 0.1)
y = np.sin(x)

plt.plot(x,y)
plt.title('Passes the vertical line test')
plt.show()
```

```python
plt.plot(y,x)
plt.title('Does not pass the vertical line test')
plt.show()
```

## 1.2 Why are graphs that don't pass the vertical line test not considered "functions?"

Because multiple outputs are mapped to the same input. Thus, it is ambigious which output should correspond with the input.

## 4.2 Verify that your transformation matrices are correct by choosing an input matrix and calculating the result both via the traditional functions above and also via vector-matrix multiplication.

```python
# with transformation matrix
pT = np.array([[1,3],[-1,2]])
ptest = np.array([[2],[1]])
result = np.matmul(pT, ptest)
result
```

```python
# without transformation matrix
result = np.array([[2+3*1],[2*1-2]])
result
```

# Eigenvalues and Eigenvectors

## 5.1 In your own words, give an explanation for the intuition behind eigenvalues and eigenvectors.

Eigenvectors are vectors that are not 'knocked off of their span' by a linear transformation. Eigenvectors may be stretched or shrunk by a scalar but do not alter their span. 

A eigenvalue is simply the scalar that shrinks or stretches the eigenvector during the linear transformation. The eigenvalue can be positive or negative. A negative eigenvalue indicates that the vector reverses direction (while retaining the same span). 

# The Curse of Dimensionality

## 6.1 What are some of the challenges of working with high dimensional spaces?
- **Sparcity:** As dimensions are added to datasets, the data points become more spread out. Given enough dimensions this can cause clusters/relationships to diminish as all datapoints become space somewhat evenly.   
- **Computational Intensive:** Running computational tasks across a dataset with a high number of dimensions inherently takes longer/uses more compute resources than low number of dimensions (assuming the same number of observations).
- **Gaining an intuition for the space:** High deimensional spaces make it difficult to explore and discover relationships between dimensions. Thus, it increases the importance of intuition and hypothesis testing.

## 6.2 What is the rule of thumb for how many observations you should have compared to parameters in your model?

- You should have at least 5 times as many observations as you do parameters
- Having more observations is almsot always better
- If the number of parameters is ever equal or greater than the number of observations you are at very grave risk of overfitting

# Principal Component Analysis

## 7.1 Load the UCI Machine Learning Repository's [Iris Dataset](https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv) and use PCA to isolate the dataset's first and second principal components and plot them on a graph. 

```python
import pandas as pd
from sklearn.decomposition import PCA
```

```python
url = 'https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.shape
```

```python
x = df.drop(columns='species')
```

### Using Sklearn

```python
pca = PCA(2)

pca.fit(x)

print('Eigenvectors: \n', pca.components_)

print('Eigenvalues: \n', pca.explained_variance_)

B = pca.transform(x)
print('Projected Data: \n', B[:5])
```

```python
for i in B:
    plt.scatter(i[0],i[1])
```
