# Vertical Line Test

## 1.1 Create two graphs, one that passes the vertical line test and one that does not.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return t+t**2+1

t = np.arange(-5,5,0.1)

plt.axis([-5, 5, 0, 5])
plt.grid(True)
plt.plot(t,f(t), 'b')
plt.plot()

plt.title("Passes Vertical Line Test")
plt.show();
```

```python
circle = plt.Circle((5, 5), 4, color='b', fill=False)

ax = plt.gca()
ax.cla()

ax.set_xlim((0, 10))
ax.set_ylim((0, 10))
plt.grid(True)

plt.title("Does Not Pass Vertical Line Test")
ax.add_artist(circle);
```

## 1.2 Why are graphs that don't pass the vertical line test not considered "functions?"

They are not considered functions because for the same set of inputs, it has two or more different outputs, which violates the rule of a what is considered a function.

# Eigenvalues and Eigenvectors

## 5.1 In your own words, give an explanation for the intuition behind eigenvalues and eigenvectors.

Eigenvectors refer to the vectors that remained unchanged in orientation or direction throughout the transformation of space. They can, however, be scaled so the eigenvalue refers to the scalar that represents how the eigenvector is scaled during the transformation.

# The Curse of Dimensionality

## 6.1 What are some of the challenges of working with high dimensional spaces?

Some of the challenges are that as you add more features, the less likely they are to contribute in a meaningful way to the data set. As you add more features, it also becomes more computationally expensive to process due to increased time. Also, measures of distance start to lose meaning since the points seem equidistant to each other. The more features you have in relation to the observations, the greater risk you incur of the model overfitting the data.

## 6.2 What is the rule of thumb for how many observations you should have compared to parameters in your model?

The rule of thumb is that you should have at least 5 times as many observations as you do parameters.

# Principal Component Analysis

## 7.1 Load the UCI Machine Learning Repository's [Iris Dataset](https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv) and use PCA to isolate the dataset's first and second principal components and plot them on a graph. 

```python
import pandas as pd
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv')
print(df.shape)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
import seaborn as sns
sns.pairplot(df, hue='species');
```

```python
df_number = df.select_dtypes(exclude=['object'])
```

```python
df_number = df_number - df_number.mean()
df_number.head()
```

```python
z_matrix = df_number/df_number.std()
```

```python
z_transpose = z_matrix.T
```

```python
z_matrix.T.dot(z_matrix)
```

```python
from numpy import cov
cov(z_transpose)
```

```python
covariance_matrix = z_matrix.cov()
covariance_matrix
```

```python
from numpy.linalg import eig
values, vectors = eig(covariance_matrix)
print("\n Eigenvectors: \n", vectors)
print("\n Eigenvalues: \n", values)
```

```python
P = vectors.T.dot(z_matrix.T)
print("\n Projected Data: \n", P.T)
```

```python
projected = pd.DataFrame(P.T)
projected.head()
```

```python
projected['species'] = df['species']
projected.head()
```

```python
sns.scatterplot(x =0, y=1, data=projected, hue='species');
```

```python
sns.pairplot(projected, hue='species');
```

```python
from sklearn.decomposition import PCA
X = df_number.values
print("Data: \n", X)
# create the PCA instance
pca = PCA(4)
# fit on data
pca.fit(X)
# access values and vectors
print("\n Eigenvectors: \n", pca.components_)
print("\n Eigenvalues: \n",pca.explained_variance_)
# transform data
B = pca.transform(X)
print("\n Projected Data: \n", B)
```

```python
new = pd.DataFrame(B)
new = new.rename(columns={0:'PCA1',1:'PCA2',2:'PCA3',3:'PCA4'})
new['species'] = df['species']
new.head()
```

```python
sns.pairplot(new, hue='species');
```

```python
sns.scatterplot(x = 'PCA1', y='PCA2', data=new, hue='species');
```
