# Vertical Line Test

## 1.1 Create two graphs, one that passes the vertical line test and one that does not.

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Passes the vertical line test

```
x = np.array(range(50))
y = x**3

plt.plot(x, y, color='red')
plt.show()
```

DOES NOT pass the vertical line test

```
radii = np.linspace(0, 2*np.pi, 100)

r = np.sqrt(1.0)

x1 = r*np.cos(radii)
x2 = r*np.sin(radii)

plt.plot(x1, x2)
plt.xlim(-1.25,1.25)
plt.ylim(-1.25,1.25)
plt.show();
```

## 1.2 Why are graphs that don't pass the vertical line test not considered "functions?"

In order for a function to be so, each input must have a unique output. The vertical line test is a simple way to check if any single x value in a graph corresponds to a single y value. If it does not, then the graph is not representative of a function

# Functions as Relations

## 2.1 Which of the following relations are functions? Why?

\begin{align}
\text{Relation 1: } \{(1, 2), (3, 2), (1, 3)\}
\\
\text{Relation 2: } \{(1, 3), (2, 3), (6, 7)\}
\\
\text{Relation 3: } \{(9, 4), (2, 1), (9, 6)\}
\\
\text{Relation 4: } \{(6, 2), (8, 3), (6, 4)\}
\\
\text{Relation 5: } \{(2, 6), (2, 7), (2, 4)\}
\end{align}

*******
Relations 2 and 3 are both functions, because each observed input has a unique output. Relations 1, 4, and 5 are not functions, because there multiple instances of a single input yielding different outputs.

```
import os.path
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import pandas as pd

# Read National Data 
national_url = 'https://www.huduser.gov/portal/datasets/hads/hads2013n_ASCII.zip'
national_file = 'thads2013n.txt'

if os.path.exists(national_file):
    national = pd.read_csv(national_file)
else: 
    z_national = urlopen(national_url)
    zip_national = ZipFile(BytesIO(z_national.read())).extract(national_file)
    national = pd.read_csv(zip_national)
```
