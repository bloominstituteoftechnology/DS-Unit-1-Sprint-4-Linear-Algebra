# Vertical Line Test

## 1.1 Create two graphs, one that passes the vertical line test and one that does not.

```
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')

fig= plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 100)
xl=np.ones(100)
ax.plot(x, np.sin(x))
ax.plot(xl, x)
plt.arrow(8,-1,0,10, linewidth=0.8, color='orange')
plt.show();
```

```
plt.style.use('seaborn-whitegrid')

fig= plt.figure()
ax = plt.axes()
x = np.linspace(-10, 10, 1000)

ax.plot(x, x**2)

plt.arrow(5,-3,0,100, linewidth=0.8, color='orange')
plt.show();
```

# Principal Component Analysis

## 7.1 Load the UCI Machine Learning Repository's [Iris Dataset](https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv) and use PCA to isolate the dataset's first and second principal components and plot them on a graph. 

```
import pandas as pd

url='https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv'
data=pd.read_csv(url)
data.head()
```

```
# x_sepal=data['sepal_length']
# y_sepal=data['sepal_width']
# x_petal=data['petal_length']
# y_petal=data['petal_width']
```

```
centered=data
centered.head()
```

```
from numpy import array
from numpy import mean
from numpy import std
from numpy import cov
from numpy.linalg import eig

means_columns=mean(data,axis=0)
print('means of the columns:',means_columns)

names=['sepal_length','sepal_width' ,'petal_length' ,'petal_width' ]
for x in names:
  centered[x] = centered[x]-means_columns[x]
```

```
centered.head() # Centered Data
```

```
# calculate the standard deviation of each column
std_devs=std(data,axis=0)
print('standard deviatiosn:',std_devs)
```

```
Z=data
for x in names:
  Z[x] = centered[x]/std_devs[x]
Z.head()#Standarized data
```

```
Z1=Z.drop(['species'],axis=1)
```

```
Z1.dtypes
```

```
# the covaraince matrix 
covariance_matrix = cov(Z1.T)
covariance_matrix
```

```
# eigendecomposition of covariance matrix
val,vect=eig(covariance_matrix) 
print(val)
print(vect)
```

```
# project data
P=(vect.T.dot(Z1.T))
print(P.T)
```

```
from numpy import array
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

pca = PCA(2)
# fit on data
pca.fit(Z1)
# access values and vectors
print("\n Eigenvectors: \n", pca.components_)
print("\n Eigenvalues: \n",pca.explained_variance_)
# transform data
B = pca.transform(Z1)
print("\n Projected Data: \n", B)
```

```
import seaborn as sns
df_final=pd.concat([(pd.DataFrame(B,columns=['sepal_length','sepal_width'])),data['species']],axis=1)
df_final.head()
```

```
# AFTER Standarized Data
ax=sns.scatterplot(data=df_final,x='sepal_length',y='sepal_width',hue='species')
```

```
sns.pairplot(df_final, hue="species");
```
