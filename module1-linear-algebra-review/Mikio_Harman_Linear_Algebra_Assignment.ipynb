
## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12,11],
              [7,10]])

J = np.array([[0,1,2],
              [7,10,4],
              [3,2,0]])

detH = np.linalg.det(H)
detJ = np.linalg.det(J)

print(detH)
print(detJ)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
h_inverse = np.linalg.inv(H)
j_inverse = np.linalg.inv(J)

print(h_inverse)
print()
print(j_inverse)
```

$HH^{-1} \neq J^{-1}J$ since they have different dimmensions and different values

```
hh = np.matmul(H, h_inverse)
jj = np.matmul(j_inverse, J)

print(hh)
print()
print(jj)
```
