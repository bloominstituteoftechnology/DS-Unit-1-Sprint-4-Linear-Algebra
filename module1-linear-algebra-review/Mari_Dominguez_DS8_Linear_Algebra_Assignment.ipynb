## 1.6 Find $c \cdot d$

\begin{align}
\vec{c} = \begin{bmatrix}7 & 22 & 4 & 16\end{bmatrix}
\qquad
\vec{d} = \begin{bmatrix}12 & 6 & 2 & 9\end{bmatrix}
\end{align}


```
c=[7,22,4,16]
d=[12,6,2,9]

def dotprod (c,d):
  sums=0
  assert len(c)==len(d)
  for i in range (len(c)):
    sums=sums+(np.multiply(c[i],d[i]));
  return sums;

cdotd=dotprod(c,d)
print(cdotd)

dot=np.dot(c,d)
print(dot)
    
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
def magnitude (g):
  sums=0
  for i in range(len(g)): 
    sums=sums+pow((g[i]),2);
  final=pow(sums,.5)
  return final;

g=[1,1,1,8]

print(magnitude(g))

print(np.linalg.norm(g))
    
```
