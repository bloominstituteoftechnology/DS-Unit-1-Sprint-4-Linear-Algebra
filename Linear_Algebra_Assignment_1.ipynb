## 1.1 Create a two-dimensional vector and plot it on a graph

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
vectors = np.array([[0, 0, 0, 1, 2, 3]])

X, Y, Z, U, V, W = zip(*vectors)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 1.5])
ax.set_ylim([0, 2.5])
ax.set_zlim([0, 3.5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("Vector [1, 2, 3]")
plt.show()
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{e} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
vect_g = [1, 1, 1, 8]
vect_h = [3, 3, 3, 3]

norm_g = [(sum(m*m for m in vect_g))**.5]
norm_h = [(sum(m*m for m in vect_h))**.5]
if (norm_g > norm_h):
  print("Norm of", vect_g,  norm_g, " is greater than Norm of", vect_h,  norm_h)
else:
  print("Norm of", vect_h,  norm_h, " is greater than Norm of", vect_g,  norm_g)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H = np.array([[12, 11],
              [2, 10]])

J = np.array([[0, 1, 2],
              [7, 10, 4],
              [3, 2, 0]])

print("Determinant of:\n", H, "\n is", int(np.linalg.det(H)))
print("\nDeterminant of:\n", J, "\n is", int(np.linalg.det(J)))
```

## 3.3 Find H^{-1} and then find J^{-1}

```
inv_H = np.linalg.inv(H)
inv_J = np.linalg.inv(J)

print("Inverse of:\n", H, "\n is:\n", inv_H)
print("\nInverse of:\n", J, "\n is:\n", inv_J)
```
