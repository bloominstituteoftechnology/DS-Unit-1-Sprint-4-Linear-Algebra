## 1.2 Create a three-dimensional vector and plot it on a graph


```
from mpl_toolkits.mplot3d import Axes3D
three_d = np.array([[1,1,1,3,3,3],
                  [1,1,1,2,0,4],
                  [1,1,1,5,4,7]])

X, Y, Z, U, V, W = zip(*three_d)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=.8)
ax.set_xlim([0, 6])
ax.set_ylim([0, 5])
ax.set_zlim([0, 8])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show();
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{e} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
from numpy import linalg as LA

vector_g = np.array([1,1,1,8])
vector_h = np.array([3,3,3,3])

print(LA.norm(vector_g))
print(LA.norm(vector_h))
```

Norm of vector g is longer.
