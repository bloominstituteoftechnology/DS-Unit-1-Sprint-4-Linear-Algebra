## 1.1 Create a two-dimensional vector and plot it on a graph

```
fig, ax = plt.subplots(figsize=(8,6))

blue = [.5, .5]

plt.arrow(0,0, blue[0], blue[1], head_width=.05, head_length=0.05, color ='blue')

plt.xlim(0,1)          
plt.ylim(0,1)
plt.title("Two-Dimensional Vector")
plt.show()
```

## 1.2 Create a three-dimensional vector and plot it on a graph

```
fig = plt.figure()

vectors = np.array([[0, 0, 0, .5, .5, .5], 
                    [0, 0, 0, .5, .5, .5],
                    [0, 0, 0, .5, .5, .5]])

ax = fig.gca(projection='3d')
# ax = fig.add_subplot(111, projection='3d')
X, Y, Z, U, V, W = zip(*vectors)
ax.quiver(X, Y, Z, U, V, W, length=1)

ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title("Three-Dimensional Vector")
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
fig, ax = plt.subplots(figsize=(8,6))

blue = [.5, .5]
five = np.multiply(blue,5)
pi = np.multiply(blue,math.pi)
e = np.multiply(blue, -1*math.e)

plt.arrow(0,0, five[0], five[1], head_width=.05, head_length=0.05, color ='red')
plt.arrow(0,0, pi[0], pi[1], head_width=.05, head_length=0.05, color ='green')
plt.arrow(0,0, e[0], e[1], head_width=.05, head_length=0.05, color ='purple')
plt.arrow(0,0, blue[0], blue[1], head_width=.05, head_length=0.05, color ='blue')

plt.xlim(-2,4)          
plt.ylim(-2,4)
ax.axhline(y=0)
ax.axvline(x=0)
plt.title("Scaled Vectors")
plt.show()
```

## 1.9 Show that the following vectors are orthogonal (perpendicular to each other):

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 0 \\ -1  \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix}
\end{align}

```
# Two vectors whose dot product is =0 (i.e., the vectors are perpendicular) 
# are said to be orthogonal

g = np.array([1,0,-1])
h = np.array([1,math.sqrt(2),1])
np.dot(h,g)
```

```
fig, ax = plt.subplots(figsize=(8,6))

origin = [0], [0]
plt.quiver(*origin, g, h, angles='xy', scale_units='xy', color=['r','b','g'], scale=1)

plt.xlim(-1.5,1.5)          
plt.ylim(-.5,1.5)
plt.title("Orthogonal Vectors g and h")
plt.show()
```
