## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([1 , 1, 1, 8])
h = np.array([3, 3, 3, 3])
norm_g = np.linalg.norm(g)
norm_h = np.linalg.norm(h)

if norm_g > norm_h:
  print(f'norm_g = {norm_g} is longer than norm_h = {norm_h}')
else:
  print(f'norm_h = {norm_h} is longer than norm_g = {norm_g}')
```
