## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
# Making the matrices.
A_m = np.matrix([[1,2], [3,4], [5,6]])
B_m = np.matrix([[2,4,6]])
C_m = np.matrix([[9,6,3], [4,7,11]])
D_m = np.matrix([[1,0,0], [0,1,0], [0,0,1]])
E_m = np.matrix([[1,3], [5,7]])

# CD:
CD = np.dot(C_m, D_m)
print("CD =", CD)

# AE:
AE = np.dot(A_m, E_m)
print("\n AE =", AE)

# BA:
BA = np.dot(B_m, A_m)
print("\n BA =", BA)

print("\n CD is a 2x3 matrix, AE a 3x2 matrix, and BA is a 1x2 matrix. Each of these has a dimension equal to the Row and Column of their factor matrices, respectively.")
```

## 3.1 Find $IG$ (be sure to show your work) 

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```
# Id is the identity matrix, which for a 2x2 square matrix is:
Id = np.matrix([[1,0], [0,1]])

# The matrix 'G':
G_m = np.matrix([[12,11], [7,10]])

# Multiplied:
IG = np.dot(Id, G_m)
print(IG)

# Since the identity matrix is similar to multipling by 1, the new matrix "IG", is actually just the same "G" matrix. 
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
H_m = np.matrix([[12,11], [7,10]])
J_m = np.matrix([[0,1,2], [7,10,4], [3,2,0]])

Det_H_m = np.linalg.det(H_m)
Det_J_m = np.linalg.det(J_m)

print("Matrix 'H' has a determinant of {}, and Matrix 'J' has a determinant of {}.".format(Det_H_m, Det_J_m))
```

## 3.3 Find H^{-1} and then find J^{-1}

```
Inv_H_m = np.linalg.inv(H_m)
Inv_J_m = np.linalg.inv(J_m)

print(Inv_H_m, "\n", "\n", Inv_J_m)
```

## 3.4 Find $HH^{-1}$ and then find $G^{-1}G$. Is $HH^{-1} == G^{-1}G$? Why or Why not?

```
H_inv_H = np.dot(H_m, Inv_H_m)
J_inv_J = np.dot(J_m, Inv_J_m)

print(H_inv_H, "\n", "\n", J_inv_J)

# They are not equal because they have different dimensions.
```
