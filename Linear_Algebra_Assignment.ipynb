## 3.1 Find $IG$ (be sure to show your work) 

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```
G = np.array([[12,11],
              [7,10]])

IG_coef = (1/(12*10 - 11*7))

IG = IG_coef*np.array([[G.item(1,1),-1*G.item(0,1)], [-1*G.item(1,0), G.item(0,0)]])

IG_np = np.linalg.inv(G)

print("'By hand':", IG, "using np:", IG_np)
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\

3 & 2 & 0
\end{bmatrix}

```
H = np.array([[12,11], [7,10]])
J = np.array([[0,1,2], [7,10,4], [3,2,0]])

print(np.linalg.det(H))
print(np.linalg.det(J))
```
