## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}
2 & 4 & 6 \\
\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
A = np.array([[1,2], [3,4], [5,6]])
B = np.array([[2,4,6]])
C = np.array([[9,6,3], [4,7,11]])
D = np.array([[1,0,0], [0,1,0], [0,0,1]])
E = np.array([[1,3], [5,7]])

#store matrices along with string name for comparing and printing
Matrices = [['A',A],['B',B],['C',C],['D',D],['E',E]] 

#empty list to append combinations of possible matrix multiplications
combos = []


for matrix in Matrices:
  for i in range(0,len(Matrices)):
    #if named matrix is the same, skip - ie AA is invalid
    if matrix[0] == Matrices[i][0]: 
      continue
    #if number of cols  equals number of rows, append names to combos list  
    elif matrix[1].shape[1] == Matrices[i][1].shape[0]:
      combos.append(matrix[0] + Matrices[i][0])
      

print('There are ',len(combos), ' legal combinations:')      
print(combos)
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
CD = np.matmul(C,D)
AE = np.matmul(A,E)
BA = np.matmul(B,A)

matmuled  = [['CD',CD],['AE',AE],['BA',BA]]

for matted in matmuled:
  print('The shape of ',matted[0], ' is', matted[1].shape)

print('\n')

for matrix in Matrices:
  print('The shape of ',matrix[0], ' is', matrix[1].shape)
```

The output shape of the product of 2 matrices is (N,M) where N = the number of rows in the first matrix and M = the number of columns in the second matrix
