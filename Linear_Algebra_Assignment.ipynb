## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
v_1 = [2.5,1.5]
v_3 = np.multiply(5, v_1)
v_4 = np.multiply(math.pi, v_1)
v_5 = np.multiply(-math.e, v_1)

plt.arrow(0,0, v_1[0], v_1[1], head_width=.25,
          head_length=0.2, color = 'r', alpha = .9)
plt.arrow(0,0, v_3[0], v_3[1], head_width=.25,
          head_length=0.2, color = 'b', alpha = .9)
plt.arrow(0,0, v_4[0], v_4[1], head_width=.25,
          head_length=0.2, color = 'g', alpha = .9)
plt.arrow(0,0, v_5[0], v_5[1], head_width=.25,
          head_length=0.2, color = 'y', alpha = .9)

plt.title('Scaled_Vectors')
plt.axis([-10, 15, -10, 15])
plt.show()
```

The vector magnitude of the scaled vectors changes based on the scale factor relative to the original vector. The sign of the scaling factor determines the direction.

## 1.5 find $\vec{a} - \vec{b}$ and plot the result on the same graph as $\vec{a}$ and $\vec{b}$. Is there a relationship between vectors $\vec{a} \thinspace, \vec{b} \thinspace \text{and} \thinspace \vec{a-b}$

```
def subtract_vectors(c,d):
  new_vector = list(map(lambda a,b: a-b, c, d))
  return new_vector

v_ab = subtract_vectors(v_a, v_b)

plt.arrow(0,0, v_a[0], v_a[1], head_width=.15,
          head_length=0.1, color = 'r', alpha = .9)
plt.arrow(0,0, v_b[0], v_b[1], head_width=.15,
          head_length=0.1, color = 'b', alpha = .9)
plt.arrow(0,0, v_ab[0], v_ab[1], head_width=.15,
          head_length=0.1, color = 'g', alpha = .9)

plt.title('Vectors_A_B and AB')
plt.axis([0, 10, 0, 10])
plt.show()
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
mH = np.matrix('12 11;7 10')
mJ = np.matrix('0 1 2;7 10 4;3 2 0')

Hdet = np.linalg.det(mH)
Jdet = np.linalg.det(mJ)

print('|H| =',Hdet)
print('|J| =',Jdet)
```

## 3.3 Find H^{-1} and then find J^{-1}

```
mHinv = np.linalg.inv(mH)
print('mH^-1 =',mHinv)

mJinv = np.linalg.inv(mJ)
print('mJ^-1 =',mJinv)
```
