## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
g = np.array([1, 1, 1, 8])
h = np.array([3, 3, 3, 3])

# Calculate lengths
g_norm = np.linalg.norm(g)
h_norm = np.linalg.norm(h)

print('Length of G:', g_norm)
print('Length of H:', h_norm)

if g_norm > h_norm:
  print('G is longer')
else:
  print('H is longer')
```

## 2.1 What are the dimensions of the following matrices? Which of the following can be multiplied together? See if you can find all of the different legal combinations.
\begin{align}
A = \begin{bmatrix}
1 & 2 \\
3 & 4 \\
5 & 6
\end{bmatrix}
\qquad
B = \begin{bmatrix}
2 & 4 & 6 \\
\end{bmatrix}
\qquad
C = \begin{bmatrix}
9 & 6 & 3 \\
4 & 7 & 11
\end{bmatrix}
\qquad
D = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1
\end{bmatrix}
\qquad
E = \begin{bmatrix}
1 & 3 \\
5 & 7
\end{bmatrix}
\end{align}

```
# Create arrays
a = np.array([[1, 2], 
              [3, 4], 
              [5, 6]])
b = np.matrix([2, 4, 6])   # Had to to get the shape correct
c = np.array([[9, 6, 3], 
              [4, 7, 11]])
d = np.array([[1, 0, 0], 
             [0, 1, 0], 
             [0, 0, 1]])
e = np.array([[1, 3], 
             [5, 7]])

# Print dimensions of matrices
print('A dimensions:', a.shape)
print('B dimensions:', b.shape)
print('C dimensions:', c.shape)
print('D dimensions:', d.shape)
print('E dimensions:', e.shape)
```

```
# Find all different "legal" combinations

# Create dictionary with matrix variable as the key, and its shape as the values
shape_dict = {'a': a.shape, 'b': b.shape, 'c': c.shape, 
             'd': d.shape, 'e': e.shape}

# Empty list for appending our combinations
pairs = []

for k, v in shape_dict.items():          # iterate through each key, value pair
  for key, value in shape_dict.items():  # iterate a second time to get all possible combinations
    if (v[0] == value[1]) & (k != key):  # check to see if columns of first matrix match rows of second, excluding multiplying the matrix by itself
      pairs.append((k, key))             # append conditional pairs to our empty list
      
pairs
```

## 2.2 Find the following products: CD, AE, and BA. What are the dimensions of the resulting matrices? How does that relate to the dimensions of their factor matrices?

```
cd = np.matmul(c, d)
ae = np.matmul(a, e)
ba = np.matmul(b, a)

print('C & D shapes:', c.shape, d.shape)
print('CD:', cd, 'CD shape:', cd.shape)
print('\n')
print('A & E shapes:', a.shape, e.shape)
print('AE:', ae, 'AE shape:', ae.shape)
print('\n')
print('B & A shapes:', b.shape, a.shape)
print('BA:', ba, 'BA shape:', ba.shape)
```

## 3.1 Find $IG$ (be sure to show your work) 

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\end{align}

```
g = np.array([[12, 11], 
             [7, 10]])

i = np.identity(2)

ig = np.matmul(i, g)
ig
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```
h = g
j = np.array([[0, 1, 2], 
             [7, 10, 4], 
             [3, 2, 0]])

h_det = np.linalg.det(h)
j_det = np.linalg.det(j)

print('H det:', h_det)
print('J det:', j_det)
```

## 3.3 Find $H^{-1}$ and then find $J^{-1}$

```
h_inv = np.linalg.inv(h)
j_inv = np.linalg.inv(j)

print('H inverse: \n', h_inv)
print('J inverse: \n', j_inv)
```

## 3.4 Find $HH^{-1}$ and then find $J^{-1}J$. Is $HH^{-1} == J^{-1}J$? Why or Why not?

```
h_h_inv = np.matmul(h, h_inv)
j_j_inv = np.matmul(j_inv, j)

print('HH^-1: \n', h_h_inv)
print('JJ^-1: \n', j_j_inv)

print('Are they equal?', h_h_inv == j_j_inv)
```

Not equal. Mainly because they're of different dimensionality, but also due to how python interprets the numbers
