## 1.1 Create a two-dimensional vector and plot it on a graph

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
from math import e, sqrt
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
vector_3d = [1, 2, 3]

fig = plt.figure()
ax = fig.gca(projection = '3d')

x, y, z = np.meshgrid(0,0,0)
u = vector_3d[0]
v = vector_3d[1]
w = vector_3d[2]

ax.set_xlim([-1, 5])
ax.set_ylim([-1, 5])
ax.set_zlim([-1, 5])
ax.quiver(x, y, z, u, v, w, color = 'purple', arrow_length_ratio = 0.2)

plt.show()
```

```
A = np.matrix([[1,2],[3,4],[5,6]])
print("The dimension of A is", A.shape)

B = np.array([[2,4,6]])
print("The dimension of B is", B.shape)

C = np.matrix([[9,6,3],[4,7,11]])
print("The dimension of C is", C.shape)

D = np.matrix([[1,0,0],[0,1,0], [0,0,1]])
print("The dimension of D is", D.shape)

E = np.matrix([[1,3],[5,7]])
print("The dimension of E is", E.shape)

#you can multiply AC, CD, BA, CA, AE, CE, DA
np.matmul(A,C)
np.matmul(C,D)
np.matmul(B,A)
np.matmul(E,C)
np.matmul(D,A)
```
