## 1.9 Show that the following vectors are orthogonal (perpendicular to each other):

\begin{align}
\vec{g} = \begin{bmatrix} 1 \\ 0 \\ -1  \end{bmatrix}
\qquad
\vec{h} = \begin{bmatrix} 1 \\ \sqrt{2} \\ 1 \end{bmatrix}
\end{align}

```python
# If their dot product is 0, they are indeed orthogonal. 
# If their dot product is 1, they are parallel.

new_array_g = [1, 0, -1]
new_array_h = [1, 1**0.5, 1]
new_array_gh = np.dot(new_array_g, new_array_h)
print ('their dot product is:', new_array_gh)

# so yea, orthogonal for sure
```

## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}

\begin{bmatrix}
