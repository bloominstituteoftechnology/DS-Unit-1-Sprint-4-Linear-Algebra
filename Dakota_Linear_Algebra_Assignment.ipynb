## 3.2 Find $|H|$ and then find $|J|$.

\begin{align}
H= 
\begin{bmatrix}
12 & 11 \\
7 & 10 
\end{bmatrix}
\qquad
J= 
\begin{bmatrix}
0 & 1 & 2 \\
7 & 10 & 4 \\
3 & 2 & 0
\end{bmatrix}
\end{align}


```python
helo= np.mat('12, 11 ; 7, 10 ')
jay= np.mat(' 0, 1, 2 ; 7, 10, 4 ; 3, 2, 0')
dh= np.linalg.det(helo)
jh = np.linalg.det(jay)
print('Determinant of H \n', dh)
print( 'Determinant of J \n',jh  )
```

## 3.3 Find H^{-1} and then find J^{-1}

```python
hin = np.linalg.inv(helo)
jin= np.linalg.inv(jay)
print('H^ -1 = \n',hin)
print('J^-1 = \n',jin)
```
