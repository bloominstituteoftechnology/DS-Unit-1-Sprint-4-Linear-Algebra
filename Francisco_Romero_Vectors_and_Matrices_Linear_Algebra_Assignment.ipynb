## 3.1 Find $IG$ (be sure to show your work) 

\begin{align}
G= 
\begin{bmatrix}
12 & 11 \\

\end{bmatrix}

```
from numpy.linalg import inv
```

```
G = np.array([[12,11],[7,10]])
G
```

```
I = np.eye(2)
I
```

```
GGT = np.dot(G.T,G)
GGT
```

```
np.linalg.det(GGT)
```

```
np.dot(np.linalg.inv(GGT),GGT)
```

```
np.allclose(np.dot(np.linalg.inv(GGT),GGT), np.eye(2))
```
