## 1.1 Create a two-dimensional vector and plot it on a graph

```
# Scaling vectors with Scalars
import math
import matplotlib.pyplot as plt
import numpy as np

# Start with a simple vector
blue = [3, 1.37]

plt.arrow(0,0, blue[0], blue[1],head_width=.05, head_length=0.05, color ='blue')

plt.xlim(-1,4)          
plt.ylim(-1,4)
plt.title("Scaled Vectors")
plt.show()
```

## 1.2 Create a three-dimensional vecor and plot it on a graph

```
# Add a third component
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

vector = np.array([[0, .3, .2, .45, .65, .5]])

X, Y, Z, U, V, W = zip(*vector)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=1)
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
# Scaling vectors with Scalars
import math
import matplotlib.pyplot as plt
import numpy as np

# Start with a simple vector
blue = [3, 1.37]
multby1 = np.multiply(1.1, blue)
multby5 = np.multiply(5, blue)
multbypi = np.multiply(np.pi, blue)
multbye = np.multiply(np.exp(blue), blue)

plt.arrow(0,0, blue[0], blue[1],head_width=.05, head_length=0.05, color ='blue')
plt.arrow(0,0, multby1[0], multby1[1],head_width=.05, head_length=0.05, color ='green')
plt.arrow(0,0, multby5[0], multby5[1],head_width=.05, head_length=0.05, color ='red')
plt.arrow(0,0, multbypi[0], multbypi[1],head_width=.05, head_length=0.05, color ='orange')
plt.arrow(0,0, multbye[0], multbye[1],head_width=.05, head_length=0.05, color ='pink')

plt.xlim(-1,100)          
plt.ylim(-1,100)
plt.title("Scaled Vectors")
plt.show()
```

the scaled vectors have the same direction except for the one scaled -e.
