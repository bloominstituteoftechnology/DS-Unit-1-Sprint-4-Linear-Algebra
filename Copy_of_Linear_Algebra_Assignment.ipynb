```
#dot product of any two vectors
#input must be numpy arrays

def vect_dot(vector1,vector2):
  vect1 = vector1[0]
  vect2 = vector2[0]
  n = len(vector1)
  answer_list = []
  for i in range(0,n):
    term = vect1[i]*vect2[i]
    answer_list.append(term)
  return sum(answer_list)
```
