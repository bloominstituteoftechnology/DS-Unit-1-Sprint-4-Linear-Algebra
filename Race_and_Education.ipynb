```
def confidence95(data):
  
  CE = np.sqrt([(data)*(data)/df.shape[0]])
  
  upper = data + (1.96*CE)
  lower = data - (1.96*CE)
  
  return print("We can state with 95% confidence that the average debt is between", lower, "and",  upper)
```
