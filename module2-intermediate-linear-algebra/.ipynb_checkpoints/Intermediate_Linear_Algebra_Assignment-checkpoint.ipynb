# Statistics

## 1.1 Sales for the past week was the following amounts: [3505, 2400, 3027, 2798, 3700, 3250, 2689]. Without using library functions, what is the mean, variance, and standard deviation of of sales from last week? (for extra bonus points, write your own function that can calculate these two values for any sized list)

```python
sales_wk = [3505, 2400, 3027, 2789, 3700, 3250, 2689]

def mean(lst):
    ret = 0
    index = 0
    for elem in lst:
        ret += elem
        index += 1
    return (ret/index)

def variance(lst):
    ret = 0
    index = 0
    x = mean(lst)
    for elem in lst:
        ret += (elem - x)**2
        index += 1
    return (ret/index)

def stdev(lst):
    return (variance(lst) ** .5)

sales_stats = [mean(sales_wk), stdev(sales_wk), variance(sales_wk)]
```

## 1.2 Find the covariance between last week's sales numbers and the number of customers that entered the store last week: [127, 80, 105, 92, 120, 115, 93] (you may use librray functions for calculating the covariance since we didn't specifically talk about its formula)

```python
import numpy as np

customers_wk = [127, 80, 105, 92, 120, 115, 93]
customer_stats = [mean(customers_wk), stdev(customers_wk), variance(customers_wk)]


np.cov(customers_wk, sales_wk, ddof=0)
```

## 1.3 Find the standard deviation of customers who entered the store last week. Then, use the standard deviations of both sales and customers to standardize the covariance to find the correlation coefficient that summarizes the relationship between sales and customers. (You may use library functions to check your work.)

```python
def standardize(lst):
    x = mean(lst)
    s = stdev(lst)
    return ([(elem - x) / s for elem in lst])
#In statistics, standardization is the process of putting different variables on the same scale
#You subtract the mean and divide by the standard deviation.

customers_std = standardize(customers_wk)
sales_std = standardize(sales_wk)

print(np.cov(customers_std, sales_std, ddof = 0))
print(customer_stats)
print(sales_stats)
```
