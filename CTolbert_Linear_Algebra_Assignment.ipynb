## 1.1 Create a two-dimensional vector and plot it on a graph

## 1.2 Create a three-dimensional vector and plot it on a graph


```
from mpl_toolkits.mplot3d import Axes3D
three_d = np.array([[1,1,1,3,3,3],
                  [1,1,1,2,0,4],
                  [1,1,1,5,4,7]])

X, Y, Z, U, V, W = zip(*three_d)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=.8)
ax.set_xlim([0, 6])
ax.set_ylim([0, 5])
ax.set_zlim([0, 8])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show();
```

## 1.3 Scale the vectors you created in 1.1 by $5$, $\pi$, and $-e$ and plot all four vectors (original + 3 scaled vectors) on a graph. What do you notice about these vectors? 

```
# Multiply by a scalar
red = np.multiply(5, two_d)
green = np.multiply(math.pi, two_d)
orange = np.multiply(-1*(math.e), two_d)

#Plot the scaled vectors
plt.arrow(0,0, two_d[0],two_d[1], head_width=.5, head_length=.5, color='blue')
plt.arrow(0,0, red[0],red[1], head_width=.5, head_length=.5, color='red')
plt.arrow(0,0, green[0],green[1], head_width=.5, head_length=.4, color='green')
plt.arrow(0,0, orange[0],orange[1], head_width=.8, head_length=.6, color='orange')
plt.xlim(-8,15)
plt.ylim(-15,30)
plt.show();
```

## 1.8 Find $||g||$ and then find $||h||$. Which is longer?

\begin{align}
\vec{e} = \begin{bmatrix} 1 \\ 1 \\ 1 \\ 8 \end{bmatrix}
\qquad
\vec{f} = \begin{bmatrix} 3 \\3 \\ 3 \\ 3 \end{bmatrix}
\end{align}

```
from numpy import linalg as LA

vector_g = np.array([1,1,1,8])
vector_h = np.array([3,3,3,3])

print(LA.norm(vector_g))
print(LA.norm(vector_h))
```
